```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib as mpl

# Ensure text is saved as editable text in SVG
mpl.rcParams['svg.fonttype'] = 'none'
# Set global font to Arial
mpl.rcParams['font.family'] = 'Arial'

# Load the CSV file (replace with your filename)
df = pd.read_csv("W:\\Users\\Daphne\\WT_RESULTS\\WT_OA\\PM_bb_distances.csv")

# Ensure the column exists
if "Average BB distance unxpanded (um)" in df.columns:
    plt.figure(figsize=(5, 6))

    # Create a boxplot in black/white
    sns.boxplot(
        y="Average BB distance unxpanded (um)",
        data=df,
        color="white",
        linewidth=1,
        boxprops=dict(edgecolor="black"),
        whiskerprops=dict(color="black"),
        capprops=dict(color="black"),
        medianprops=dict(color="black"),
    )

    # Overlay datapoints (as circles)
    sns.stripplot(
        y="Average BB distance unxpanded (um)",
        data=df,
        size=5,
        color="black",
        edgecolor="black",
        jitter=True,
    )

    # Titles and labels
    plt.title("Average BB Distance", fontsize=24)
    plt.xlabel(None)
    plt.ylabel("Distance (um)", fontsize=20)
    plt.ylim(0.2, 0.6)
    plt.xticks(fontsize=18)
    plt.yticks(fontsize=18)
    # Add tick marks to the y-axis
    plt.tick_params(axis='y', direction='in', length=6, width=1, color='black')
    plt.tight_layout()

    # Set background to white and add a black border
    ax = plt.gca()
    ax.set_facecolor("white")
    for spine in ax.spines.values():
        spine.set_edgecolor("black")
        spine.set_linewidth(1)

    plt.savefig("W:\\Users\\Daphne\\WT_RESULTS\\WT_OA\\PM_bb_distances_plot.svg", dpi=300)
    plt.savefig("W:\\Users\\Daphne\\WT_RESULTS\\WT_OA\\PM_bb_distances_plot.png", dpi=300)

    plt.show()

else:
    print("The CSV must contain 'Average BB distance unxpanded (um)' column.")
```

```python
import pandas as pd

# Step 1: Read the CSV file
file_path = 'W:\\Users\\Daphne\\WT_RESULTS\\WT_OA\\PM_bb_distances.csv'
df = pd.read_csv(file_path)

# Step 2: Extract the column
column_name = 'Average BB distance unxpanded (um)'
data = df[column_name]

# Step 3: Calculate mean and standard deviation
mean_value = data.mean()
std_value = data.std()

# Step 4: Print results
print(f"Mean: {mean_value:.4f}")
print(f"Standard Deviation: {std_value:.4f}")
```
