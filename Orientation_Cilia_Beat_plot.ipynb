```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file (replace with your filename)
df = pd.read_csv("W:\\Users\\Daphne\\WT_RESULTS\\WT_CB_direction\\CB_direction_allresults_filtered_regions.csv")

# Ensure columns exist
if "Region" in df.columns and "std_angle" in df.columns:
    fig = plt.figure(figsize=(5, 6))
    print(fig.get_size_inches())  # Verify the size in inches

    # Create a boxplot in black/white
    box = df.boxplot(
        column="std_angle",
        by="Region",
        grid=False,
        patch_artist=False,  # no color fill
        boxprops=dict(color="black"),
        whiskerprops=dict(color="black"),
        capprops=dict(color="black"),
        medianprops=dict(color="black"),
        flierprops=dict(marker='o', markerfacecolor='white', markeredgecolor='black', markersize=5)
    )

    # Overlay datapoints (as circles)
    regions = df["Region"].unique()
    for i, region in enumerate(sorted(regions), start=1):
        y = df.loc[df["Region"] == region, "std_angle"]
        x = [i] * len(y)
        plt.scatter(x, y, color="none", s=20, edgecolors="black")

    # Titles and labels
    plt.title(None)
    plt.suptitle("")  # remove automatic suptitle
    plt.xlabel(None)
    plt.ylabel("angle with AP axis (degrees)", fontsize=12)
    # plt.gca().set_facecolor("white")  # ensure white background
    plt.tight_layout()

    plt.savefig("boxplot_std_angle_by_region_bw_regions.png", dpi=300)
    plt.savefig("boxplot_std_angle_by_region_bw_regions.svg", dpi=300)

    plt.show()

else:
    print("The CSV must contain 'Region' and 'std_angle' columns.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib as mpl

# Ensure text is saved as editable text in SVG
mpl.rcParams['svg.fonttype'] = 'none'
# Set global font to Arial
mpl.rcParams['font.family'] = 'Arial'


# Load the CSV file (replace with your filename)
df = pd.read_csv("W:\\Users\\Daphne\\WT_RESULTS\\WT_OA\\PM_CBdirection\\PM_CB_direction_allresults.csv")

# Ensure columns exist
if "Region" in df.columns and "std_angle" in df.columns:
    plt.figure(figsize=(5, 6))

    # Create a boxplot in black/white
    sns.boxplot(
        x="Region",
        y="std_angle",
        data=df,
        color="white",
        linewidth=1,
        boxprops=dict(edgecolor="black"),
        whiskerprops=dict(color="black"),
        capprops=dict(color="black"),
        medianprops=dict(color="black"),
    )

    # Overlay datapoints (as circles)
    sns.stripplot(
        x="Region",
        y="std_angle",
        data=df,
        size=5,
        color="black",
        edgecolor="black",
        jitter=True,
    )

    # Titles and labels
    plt.title("Cilia beat direction", fontsize=24)
    plt.xlabel(None)
    plt.ylabel("angle with AP axis (degrees)", fontsize=20)
    plt.xticks(fontsize=18)
    plt.yticks(fontsize=18)
    plt.tight_layout()

    plt.savefig("boxplot_std_angle_by_region_bw_regions_sns.png", dpi=300)
    plt.savefig("boxplot_std_angle_by_region_bw_regions_sns.svg", dpi=300)

    plt.show()

else:
    print("The CSV must contain 'Region' and 'std_angle' columns.")

```

```python
import numpy as np

plt.figure(figsize=(5, 6))

sns.boxplot(
    x="Region",
    y=np.abs(df["std_angle"]),
    data=df.assign(std_angle_abs=np.abs(df["std_angle"])),
    color="white",
    linewidth=1,
    boxprops=dict(edgecolor="black"),
    whiskerprops=dict(color="black"),
    capprops=dict(color="black"),
    medianprops=dict(color="black"),
)

sns.stripplot(
    x="Region",
    y=np.abs(df["std_angle"]),
    data=df.assign(std_angle_abs=np.abs(df["std_angle"])),
    size=5,
    color="black",
    edgecolor="black",
    jitter=True,
)

plt.title("Cilia beat direction (absolute)", fontsize=24)
plt.xlabel(None)
plt.ylabel("absolute angle with AP axis (degrees)", fontsize=20)
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
plt.ylim(0, 90)
plt.tight_layout()

plt.savefig("boxplot_abs_std_angle_by_region_bw_regions_sns.png", dpi=300)
plt.savefig("boxplot_abs_std_angle_by_region_bw_regions_sns.svg", dpi=300)

plt.show()
```

```python
import pandas as pd
import numpy as np

# Load the CSV file
df = pd.read_csv(r'W:\Users\Daphne\WT_RESULTS\WT_OA\PM_CBdirection\PM_CB_direction_allresults.csv')

# Compute absolute values of 'std_angle'
abs_std_angle = np.abs(df['std_angle'])

# Calculate mean and standard deviation
mean_abs = abs_std_angle.mean()
std_abs = abs_std_angle.std()

# Print results
print(f"Mean of absolute std_angle: {mean_abs:.4f}")
print(f"Standard deviation of absolute std_angle: {std_abs:.4f}")
```
